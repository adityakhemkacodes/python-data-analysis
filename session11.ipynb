```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
df1= pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
plt.figure(figsize=(10,8))
data= df.groupby(['venue','id']).count().droplevel(level=1).index.value_counts()
sns.barplot(y=data.index, x= data, orient='h')
plt.xlabel('Matches played')
plt.ylabel('Venue')
plt.show()
```

```python
df1['bowling_team'].value_counts().sort_values(ascending= False)
```

```python
plt.figure(figsize=(10,8))
data= df1['bowling_team'].value_counts().sort_values(ascending= False)
sns.barplot(y=data.index, x= data, orient='h')
plt.xlabel('Matches played')
plt.ylabel('Team')
plt.show()
```

```python
df1.groupby(['batsman'])['batsman_runs'].sum().sort_values(ascending= False)
```

```python
plt.figure(figsize=(10,8))
data= df1.groupby(['batsman'])['batsman_runs'].sum().sort_values(ascending= False)[:15]
sns.barplot(y=data.index, x= data, orient='h')
plt.xlabel('runs')
plt.ylabel('Player')
plt.show()
```

```python
df1[df1['over']<6].groupby(['id','batting_team']).sum()['total_runs'].groupby('batting_team').mean().sort_values(ascending= False)
```

```python
runs= df1.groupby(['batsman','id'])['batsman_runs'].sum()
runs[runs >=100 ].droplevel(level=1).groupby('batsman').count().sort_values(ascending= False)[:10]
```

```python
runs= df1.groupby(['batsman','id'])['batsman_runs'].sum()
runs[runs >=50 ].droplevel(level=1).groupby('batsman').count().sort_values(ascending= False)[:10]
```

```python
plt.figure(figsize=(10,8))
runs= df1.groupby(['batsman','id'])['batsman_runs'].sum()
data= runs[runs >=50 ].droplevel(level=1).groupby('batsman').count().sort_values(ascending= False)[:10]
sns.barplot(y=data.index, x= data, orient='h')
plt.xlabel('centuries')
plt.ylabel('Player')
plt.show()
```

```python
plt.figure(figsize=(10,8))
data= df1[df1['batsman_runs']==6]['batsman'].value_counts()[:15]
sns.barplot(y=data.index, x= data, orient='h')
plt.xlabel('sixes')
plt.ylabel('Player')
plt.show()
```
